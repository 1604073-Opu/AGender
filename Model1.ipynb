```python
import os
from keras.models import load_model
import Dataset
import CNN
import matplotlib.pyplot as plt
import numpy as np
import SaveData
import tensorflow as tf
```

```python
def train():
    X_train, X_test, y_train_a, y_train_g, y_test_a, y_test_g = Dataset.load_dataset()
    model = CNN.net()
    hist = model.fit(X_train, [y_train_g, y_train_a], batch_size=86, epochs=100,
                     validation_data=(X_test, [y_test_g, y_test_a]))
    model.save('agender.h5')
    SaveData.export_model(tf.train.Saver(),'agender_age', model, ["Input"], "pred_age/Softmax")
    SaveData.export_model(tf.train.Saver(),'agender_gender', model, ["Input"], "pred_gender/Softmax")
    SaveData.saveHist('history',hist)
    return model, hist


def load():
    if os.path.exists('agender.h5'):
        model = load_model('agender.h5')
        return model,None
    else:
        model, hist = train()
        return model, hist
```

```python
model,history = load()
```

```python
X_train, X_test, y_train_a, y_train_g, y_test_a, y_test_g = Dataset.load_dataset()
```

```python
x = 26
img = X_test[x].reshape(32, 32)
ages = np.arange(0, 14).reshape(14, 1)
plt.imshow(img,cmap='gray')
true_age=np.argmax(y_test_a[x])*5
print('True: '+str(int(true_age))+'-'+str(int(true_age+4))+ 'Y, ' + ('M' if np.argmax(y_test_g[x]) == 1 else 'F'))
face = img.reshape(-1, 32, 32, 1)
pred = model.predict(face)
predicted_genders = pred[0]
predicted_ages = pred[1].dot(ages).flatten()
predicted_ages=np.argmax(pred[1])*5
g = np.argmax(predicted_genders[0])
if g == 0:
    gender = 'F'
else:
    gender = 'M'
print('Predict: '+str(int(predicted_ages))+'-'+str(int(predicted_ages+4))+'Y, '+gender)
```

```python
import matplotlib.pyplot as plt


# Plot training & validation accuracy values
plt.plot(history.history['pred_gender_acc'])
plt.plot(history.history['val_pred_gender_acc'])
plt.title('Model accuracy : Gender')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Gender_Train','Gender_Val'], loc='lower right')
plt.savefig('ACC_Gender.png')
plt.show()

plt.plot(history.history['pred_age_acc'])
plt.plot(history.history['val_pred_age_acc'])
plt.title('Model accuracy : Age')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Age_Train','Age_val'], loc='lower right')
plt.savefig('ACC_Age.png')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['pred_gender_loss'])
plt.plot(history.history['val_pred_gender_loss'])
plt.title('Model loss : Gender')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Gender_Train','Gender_Val'], loc='upper right')
plt.savefig('Loss_Gender.png')
plt.show()

plt.plot(history.history['pred_age_loss'])
plt.plot(history.history['val_pred_age_loss'])
plt.title('Model loss : Age')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Age_Train','Age_val'], loc='upper right')
plt.savefig('Loss_Age.png')
plt.show()
```
